```python
import cv2 as cv
import matplotlib.pyplot as plt

# Load the image
img = cv.imread("cubes.jpg")

# Check if the image is loaded properly
if img is None:
    print("Error: Image not found or unable to read.")
else:
    # Convert the image from BGR (OpenCV format) to RGB (Matplotlib format)
    img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    
    # Display the image using Matplotlib
    plt.imshow(img_rgb)
    plt.axis('off')  # Hide axis
    plt.show()
```

```python
# Resize the image to half its size
resized_img = cv.resize(img, (0, 0), fx=0.5, fy=0.5)

# Flip the image vertically
flipped_img = cv.flip(img, 0)

# Blur the image using Gaussian Blur
blurred_img = cv.GaussianBlur(img, (15, 15), 0)
```

```python
# Convert the image to grayscale
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# Apply simple thresholding
_, binary_img = cv.threshold(gray_img, 127, 255, cv.THRESH_BINARY)

# Detect edges using Canny Edge Detection
edges = cv.Canny(gray_img, 100, 200)
```

```python
type(edges)
```

```python
plt.imshow(edges)
```

```python
# Find contours in the image
contours, _ = cv.findContours(edges, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)

# Draw the contours on the image
cv.drawContours(img, contours, -1, (0, 255, 0), 3)
```
